## Diccionarios


```python
alumno = {
    "ciudad": "Granada",
    "edad": 32,
    "nombre": "Pepito",
}

print(alumno["nombre"])

alumno["matriculado"] = True
print(alumno)
alumno["nombre"] = "Pepe"
print(alumno)

#########

# Si intentamos acceder a una propiedad o clave sin que exista, nos arrojará un error por terminal
# print(alumno["altura"])

print(alumno.get("altura", "No existe la altura"))
print(alumno.get("nombre"))
```

```python
# METODOS DE DICCIONARIOS
alumno = {
    "nombre": "Pepito",
    "edad": 32,
    "ciudad": "Granada"
}

# Keys -> Esto devuelve una vista de todas las claves
print(alumno.keys())
# Values -> Esto devuelve una vista de todos los valores
print(alumno.values())
# Items -> Devuelve una vista de pares en tuplas de clave y valor
print(alumno.items())
# Update -> Actualizar un diccionario con otro diccionario
alumno.update({"altura": 1.90, "pelo": "rubio"})
print(alumno)
# Pop -> Le indicamos la clave y elimina la clave y valor
alumno.pop("altura")
print(alumno)
# Copy -> hacer una copia superficial de un diccionario
copia_alumno = alumno.copy()
copia_alumno["nombre"] = "Paco"
copia_alumno["edad"] = 25
print(copia_alumno)
print(alumno)
```
