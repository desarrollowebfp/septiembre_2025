```python
# break

numeros = [1,2,3,4,5,6,7]

for numero in numeros:
    print(numero)
    
    if numero == 3:
        break
    
    
alumnos = ["Jorge", "Sara", "Manuel", "Carlos", "Gabriel"]

for nombre in alumnos:
    print(f'Revisando:', nombre)
    
    if(nombre == "Sara"):
        print(f"{nombre} encontrado")
        break
    
```

```python
# continue

numeros = [1,2,3,4,5,6,7]

for numero in numeros:
    if numero == 3:
        continue
    
    print(numero)
    
    
alumnos = ["Jorge", "Sara", "Manuel", "Byron", "Carlos", "Gabriel"]

for nombre in alumnos:
    if(nombre == "Byron"):
        continue
    
    print(f"Alumno normal:", nombre)
```

```python
numeros = [1,2,3,4,5]

for numero in numeros:
    print(numero)
else:
    print("El bucle ha terminado")
```
