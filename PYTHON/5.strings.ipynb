## Strings


```python
texto = "Batman"

# Podemos indexar un string utilizando indices positivos o negativos para recorrerlo al derecho o al revés
print(texto[0])
print(texto[-1])
```

```python
# INDEXACION COMPLEJA
# string[start:stop:step]

# start -> indice de inicio
# stop -> indice final
# step -> saltos entre caracteres

personaje = "Gandalf"

# Extraer desde el 0 hasta el 3 -> el stop es excluido
print(personaje[0:3])
# Si no queremos indicar el start por obviedad podemos hacer lo siguiente
print(personaje[:3])
# Empezamos desde el 3 y omitimos el final
print(personaje[3:])
# Quiero leer el string entero pero con un salto de dos en dos caracteres
print(personaje[::2])
# Invertir un texto
print(personaje[::-1])
```

```python
# Métodos de string

# Longitud
texto = "Ash"
print(len(texto))

# Lower
print(texto.lower())

# Upper
print(texto.upper())

frase = "el caballero oscuro"
# Title -> Ponerle a cada uno de los substring la primera en mayúscula
print(frase.title())

print("Clase de programacion con Python: el tipo de dato String".title())

# Capitalize -> La primera en mayusculas
print(frase.capitalize())

# Strip -> Elimina espacios iniciales y finales
frase_con_espacios = "   Hola, buenas tardes.   "
print(len(frase_con_espacios))
frase_con_espacios = frase_con_espacios.strip()
print(len(frase_con_espacios))     

# Replace -> Reemplazar subcadenas
frase_replace = "Buenas noches Gotham!"
print(frase_replace.replace("Gotham", "Móstoles"))

# Find -> Encuentra el primer indice del string o letra coincidente
frase = "Hola, que tal"
print(frase.find("x"))

# Index -> Hace lo mismo que el find, pero si no lo encuentra arroja un error
#print(frase.index("x"))

# Join
superheroes = ["Batman", "Superman", "Wonderwoman"]
print(", ".join(superheroes))
```

```python
# Traductor a balleno
texto_original = input("Introduce un bloque de texto")
mapa = texto_original.maketrans("aeiouAEIOU", "oooooOOOOO")
texto_balleno = texto_original.translate(mapa)
print(texto_balleno)
```
