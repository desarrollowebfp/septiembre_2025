## Operadores


```python
# Operadores aritméticos
a = 11
b = 5

# Suma
print(a + b)

# Resta 
print(a - b)

# Multiplicación
print(a * b)

# División
print(a / b)

# División Entera
print(a // b)

# Módulo
print(a % b)

# Potencia
print(a ** b)
```

```python
# Operador de concatenación

saludo = "Hola"
nombre = "Ana"
edad = 30

mensaje = saludo + " " + nombre
segundo_mensaje = "Tengo " + str(edad) + " años."

print(mensaje)
print(segundo_mensaje)
```

```python
# Operadores de comparación

y = 10
z = 20

# Igual que
print(y == z)
# Desigual que
print(y != z)
# Mayor que
print(y > z)
# Mayor o igual que
print(y >= z)
# Menor que
print(y < z)
# Menor o igual que
print(y <= z)
```

```python
# Operadores lógicos
edad = 30
es_estudiante = True
tiene_carnet = False

puede_conducir = (edad >= 18) and tiene_carnet
es_joven = (edad < 28) or es_estudiante

print(puede_conducir)
print(es_joven)
```

```python
# Operadores de asignación combinados
x = 10
y = 15

# Operación y asignación
# Esto es un atajo para hacer x = x + y
x += y
x -= y
x *= y
x //= y
print(x)

saldo = 200
saldo -= 30 # gasto de comida
saldo -= 2 # gasto de pipas
saldo -= 100 # gasto del seguro
saldo += 600 # nomina
print(saldo)
```

```python
# Conversión de datos

# String > Numero Entero
numero = int("5")
print(numero, type(numero))
# String > Numero Decimal
numero_decimal = float("6.5")
print(numero_decimal, type(numero_decimal))
# Numero de cualquier tipo > Texto
texto = str(56.5)
print(texto, type(texto))
# Numero > Boolean -> Todo lo que no sea 0 exacto es True
boolean = bool(0)
print(boolean, type(boolean))
```

```python
precio = float(input("Introduce el precio del producto:"))
cantidad = int(input("Introduce la cantidad:"))

total = cantidad * precio

if cantidad > 5:
    total *= 0.95

print("Total a pagar: ", round(total, 2), "€")
print(f"{total:.2f}")
```
