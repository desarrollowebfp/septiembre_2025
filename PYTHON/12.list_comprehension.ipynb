## List Comprehension


```python
#     nueva_lista = [expresion for elemento in iterable]
#                 lo que guardas      de donde lo sacas
```

```python
numeros = []

for n in range(5):
    numeros.append(n)
print(numeros)
```

```python
numeros = [n for n in range(5)]
print(numeros)
```

```python
cuadrados = [n * n for n in range(8)]
print(cuadrados)
```

```python
cubos = [n ** 3 for n in range(8)]
print(cubos)
```

```python
pares = [n for n in range(20) if n % 2 == 0]
print(pares)
```

```python
nombres = ["anA", "daViD", "diAna"]

nombres_corregidos = [n.capitalize() for n in nombres]
print(nombres_corregidos)
```

```python
nombres = ["anA", "daViD", "diAna"]

nombres_mas_3_letras = [nombre.capitalize() for nombre in nombres if len(nombre) > 3]
print(nombres_mas_3_letras)
```

```python
numeros = [f"{n} par" if n % 2 == 0 else f"{n} impar" for n in range(20)]
print(numeros)
```

```python
tallas = ["S", "M", "L"]
colores = ["azul", "negro"]

variantes = [(t, c) for t in tallas for c in colores]
print(variantes)
```

```python
diccionario = {numero: numero * numero for numero in range(1, 11)}
print(diccionario)
```

```python
n = 10

fib = []

a = 0
b = 1

for _ in range(n):
    fib.append(a)
    a = b
    b = a + b
    
print(fib)

```

```python
n = 10
fib = [0, 1]

[fib.append(fib[-1] + fib[-2]) for _ in range(10)]
print(fib)
```
