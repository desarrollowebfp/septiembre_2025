## Condicionales


```python
espectadores = 0

if espectadores > 0:
    print("Abrir la sala")
else:
    print("No abrimos la sala")
```

```python
puntuacion = 5

if puntuacion >= 9:
    print("Obra maestra")
elif puntuacion >= 7:
    print("Muy buena")
elif puntuacion >= 5:
    print("Aceptable")
elif puntuacion >= 3:
    print("Normalita")
else:
    print("Malísima")
```

```python
edad = 15

if not edad >= 18:
    print("Entrada prohibida a menores de edad")
else:
    print("Acceso permitido")
```

```python
dia = "lunes"
hora = 17

if dia == "viernes" and hora >= 15:
    print("Sesion especial de tarde")
elif dia == "sabado" or dia == "domingo":
    print("Sesiones de fin de semana regulares")
else:
    print("Sesiones entre semana")
```

```python
dia = 3

match dia:
    case 1:
        print("Lunes")
    case 2:
        print("Martes")
    case 3:
        print("Miércoles")
    case 4:
        print("Jueves")
    case 5:
        print("Viernes")
    case 6:
        print("Sábado")
    case 7:
        print("Domingo")
    case _:
        print("No se que dia es ese")
```
