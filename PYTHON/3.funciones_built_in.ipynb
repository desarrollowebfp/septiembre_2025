## Funciones Built-In


```python
# PRINT
print("Esto es Python")

variable = 270
print(variable)

curso = "Python"
nivel = "Inicial"
minutos = 60

print("Curso:", curso, "Nivel:", nivel, "Minutos:", minutos, "min", sep="  ")

print("Esto es un texto sin salto final...", end="")
print("...Esta es la continuación")

print("""
======================      
     Demostración
====================== 

- Paso 1: Hola
- Paso 2: Adios
""")
```

```python
# TYPE

titulo = "Esto es un titulo"
numero = 25
numero_decimal = 25.5
sin_valor = None
lista = [1,2,3]
tupla = (1,2,3)
diccionario = {"clave": "valor"}

print(type(titulo))
print(type(numero))
print(type(numero_decimal))
print(type(sin_valor))
print(type(lista))
print(type(tupla))
print(type(diccionario))
```

```python
# LEN
frase = " Vamos a ver la longitud de esta frase "
print(len(frase))

frase_sin_espacios = frase.strip()
print(len(frase_sin_espacios))
```

```python
# INPUT
nombre = input("Dime tu nombre")
edad = input ("Dime tu edad")
correo = input("Dime tu correo electrónico")

print("Nombre:", nombre)
print("Edad:", edad)
print("Correo:", correo)
```

```python
#FORMAT
pelicula = "Jurassic Park"

plantilla_de_texto = """
=====================
 LA PELICULA DE HOY
=====================

- {pelicula}

_____________________
"""

plantilla_de_texto_formateada = plantilla_de_texto.format(
    pelicula=pelicula
)

print(plantilla_de_texto_formateada)
```
