## Listas


```python
lista = ["String", 24, True, None, []]

animales = ["Perro", "Gato", "Ciervo", "Loro"]

# Longitud de una lista
print(len(animales))

# Acceder a los elementos de una lista
print(animales[-1])
```

```python
# Listas anidadas
peliculas = ["Terror", ["The Evil Dead", "The Thing", "The Exorcist"], "Ciencia ficción", ["2001", "Dune", "Gravity"]]

print(peliculas[1][1])

pelicula = "Jurassic Park"
print(list(pelicula))
```

```python
# Indexación
peliculas = ["The Evil Dead", "The Thing", "The Exorcist", "2001", "Dune", "Gravity"]

# [start:stop:step]

print(peliculas[2:4])
print(peliculas[2:])
print(peliculas[::2])
print(peliculas[::-1])

# No modificamos la lista original, lo hacemos a nivel de consulta y lo podemos almacenar en una nueva variable
peliculas_al_reves = peliculas[::-1]
print(peliculas)
```

```python
# Métodos de listas

# min max 
numeros = [2,45,65,4,7,1,89,0.5,120,43,23]
print(max(numeros))
print(min(numeros))

palabras = ["Hola", "Adios", "Merienda", "Movil", "Ratón", "Impresora"]
print(max(palabras))
print(min(palabras))

# sorted - sort -> sorted ordena alfabeticamente o numericamente la lista a nivel de consulta y sort modifica la lista original
peliculas = ["The Evil Dead", "The Thing", "The Exorcist", "2001", "Dune", "Gravity"]
print(sorted(peliculas))
peliculas.sort()
print(peliculas)

numeros = [2,45,65,4,7,1,89,45,0.5,120,43,23,45]
print(sorted(numeros))

# count -> Hace un recuento de cuantas veces aparece un elemento en la lista
print(numeros.count(45))

# reverse -> Modifica el orden al inverso de una lista
print(peliculas)
peliculas.reverse()
print(peliculas)
```

```python
# Metodos para modificar los elementos de las listas

# append -> Añade un nuevo elemento al final de la lista
peliculas = ["The Evil Dead", "The Thing", "The Exorcist", "2001", "Dune", "Gravity"]

peliculas.append("El Viaje de Chihiro")
print(peliculas)

# extend -> Añade al final una lista nueva
pelis_extras = ["Spiderman", "Dracula", "Nosferatu"]
peliculas.extend(pelis_extras)
print(peliculas)

# Sustuir un elemento por otro
peliculas[3] = "Scream"
print(peliculas)

# insert -> Inserta un elemento en la lista en una posición especifica
peliculas.insert(1, "Jurassic Park")
print(peliculas)
```

```python
# Métodos de eliminación

peliculas = ["The Evil Dead", "The Thing", "The Exorcist", "2001", "Dune", "Gravity"]

# pop -> Por defecto elimina el último elemento, pero, si le pasamos un indice, elimina ese
peliculas.pop(1)
print(peliculas)

# remove -> Elimina el primer elemento coincidente con el argumento, es decir, le pasamos el elemento para que lo borre
peliculas.remove("2001")
print(peliculas)

# clear -> Vacia por completo la lista
peliculas.clear()
print(peliculas)
```
