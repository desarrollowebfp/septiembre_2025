## Funciones


```python
def saludar():
    print("Hola")
    
saludar()
saludar()
saludar()
```

```python
                     #parametro
def saludar_con_nombre(nombre):
    print(f'Hola {nombre}')
    
                  #argumento
saludar_con_nombre("Gabriel")
saludar_con_nombre("Sara")
```

```python
def sumar_dos_numeros(numeroa, numerob):
    total = numeroa + numerob
    print(total)
    
sumar_dos_numeros(5, 6)
```

```python
def multiplicar_por_dos(numero):
    total = numero * 2
    return numero * 2
    
resultado = multiplicar_por_dos(11) + multiplicar_por_dos(2)
print(resultado)
```

```python
def sumar (numeroa, numerob):
    """
    Suma dos numeros dados por argumento

    Args:
        numeroa (int, float): El primer numero de la suma
        numerob (int, float): El segundo numero de la suma

    Returns:
        int, float: Retorna el resultado de la suma
    """   
    
    total = numeroa + numerob
    return total
```

```python
# args -> parametro que recibe una cantidad indefinida/variable de valores
def mostrar_nombres(*args):
    for nombre in args:
        print("El nombre es:", nombre)
    
mostrar_nombres("Paco", "Gabriel", "Alex", "Carlos")

def sumar_numeros(*args):
    total = 0
    for numero in args:
        total += numero
    print(total)
    
sumar_numeros(1,2,3,4,5,20,12,190,45)
```

```python
#kwargs -> recibir una cantidad indefinida de claves con valores, generando así un mini-diccionario -> KEYWORDS ARGUMENTS
def mostrar_info(**kwargs):
    for clave, valor in kwargs.items():
        print(f'{clave} -> {valor}')
    
mostrar_info(nombre="Ana", edad=20, ciudad="Madrid", profesion="Carpintera")

```

```python
# lambda -> Funciones cortas reutilizables con retorno implicito
def sumar(numa, numb):
    return numa + numb

resultado_suma = lambda numa, numb: numa + numb
print(resultado_suma(2,5))
```
