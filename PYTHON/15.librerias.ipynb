```python
import sys
!pip install pygame pymunk

import pygame
import pymunk
import pymunk.pygame_util
import math

# 1. Pygame Initialization
pygame.init()
SCREEN_WIDTH, SCREEN_HEIGHT = 800, 600
screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))
pygame.display.set_caption("Hexágono y Bola con Físicas")
clock = pygame.time.Clock()

# 2. Pymunk Initialization
space = pymunk.Space()
space.gravity = (0, 900) # Gravity downwards (positive y is down in Pygame)
draw_options = pymunk.pygame_util.DrawOptions(screen)

# 3. Function to create a hexagon
def create_hexagon(space, pos, size, rotation_speed):
    # A kinematic body moves/rotates manually but interacts with dynamic bodies
    body = pymunk.Body(body_type=pymunk.Body.KINEMATIC)
    body.position = pos
    body.angular_velocity = rotation_speed # Constant angular velocity for rotation
    space.add(body) # Add the body to the space FIRST

    # Define hexagon vertices relative to its body's center
    vertices = []
    for i in range(6):
        angle = math.radians(60 * i)
        x = size * math.cos(angle)
        y = size * math.sin(angle)
        vertices.append((x, y))

    # Create segment shapes for each side of the hexagon
    shapes = []
    for i in range(6):
        # pymunk.Segment(body, vertex_start, vertex_end, radius_of_segment)
        segment = pymunk.Segment(body, vertices[i], vertices[(i + 1) % 6], 5) # 5 is the thickness of the wall
        segment.friction = 0.8 # Higher friction to prevent ball from sliding too easily
        segment.elasticity = 0.9 # High elasticity for good bounces
        segment.color = (0, 255, 0, 255) # Green color for the hexagon
        shapes.append(segment)
        space.add(segment) # Now the shapes can be added

    return body, shapes

# 4. Function to create a ball
def create_ball(space, pos, radius, mass):
    # Dynamic body: mass and inertia (moment)
    moment = pymunk.moment_for_circle(mass, 0, radius)
    body = pymunk.Body(mass, moment)
    body.position = pos
    shape = pymunk.Circle(body, radius)
    shape.friction = 0.7
    shape.elasticity = 0.95
    shape.color = (255, 0, 0, 255) # Red color for the ball
    space.add(body, shape)
    return body, shape

# Setup the hexagon
hexagon_pos = (SCREEN_WIDTH // 2, SCREEN_HEIGHT // 2)
hexagon_size = 150 # Distance from center to vertex
hexagon_rotation_speed = 0.5 # Radians per second, adjust for faster/slower rotation
hexagon_body, hexagon_shapes = create_hexagon(space, hexagon_pos, hexagon_size, hexagon_rotation_speed)

# Setup the ball, placing it inside the hexagon initially
ball_radius = 20
ball_mass = 10
# Place the ball slightly above the center of the hexagon to let gravity work
ball_pos = (SCREEN_WIDTH // 2, SCREEN_HEIGHT // 2 - hexagon_size + ball_radius * 2)
ball_body, ball_shape = create_ball(space, ball_pos, ball_radius, ball_mass)

# 5. Game Loop
running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
            # Cleanly exit if in a Colab environment or similar headless setup
            if 'google.colab' in sys.modules:
                pygame.display.quit()
                pygame.quit()
                sys.exit()

    # Physics update
    dt = 1.0 / 60.0 # Time step for 60 FPS simulation
    space.step(dt)

    # Drawing
    screen.fill((50, 50, 50)) # Dark grey background

    # Draw all objects in the Pymunk space
    space.debug_draw(draw_options)

    pygame.display.flip() # Update the full display Surface to the screen
    clock.tick(60) # Limit frame rate to 60 FPS

pygame.quit() # Uninitialize Pygame modules
```
