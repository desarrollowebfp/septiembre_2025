```python
clasicos = {"Casablanca", "Titanic", "Forrest Gump", "El Padrino", "Gladiator"}
scifi = {"Matrix", "Interstellar", "Inception", "Titanic", "Avatar"}

# Metodos basicos
# Len
print(len(clasicos))
# Add -> Añadir un elemento
scifi.add("Dune")
print(scifi)
# Update -> Añadir elementos en un listado
scifi.update(["2001", "Hail Mary"])
print(scifi)
# Remove -> Elimina un elemento especifico -> si no: Error
scifi.remove("Dune")
print(scifi)
# Discard -> Elimina un elemento especifico -> si no: No da Error
scifi.discard("Dune")
```

```python
# Métodos entre conjuntos
clasicos = {"Casablanca", "Titanic", "Forrest Gump", "El Padrino", "Gladiator"}
scifi = {"Matrix", "Interstellar", "Inception", "Titanic", "Avatar"}

# UNION -> | -> Une dos conjuntos y elimina duplicados
todos = clasicos.union(scifi)
todos = clasicos | scifi
print(todos)

# INTERSECCION -> & -> Muestra elementos que aparecen en ambos conjuntos
comunes = clasicos.intersection(scifi)
comunes = clasicos & scifi
print(comunes)

# DIFERENCIA -> - -> Muestra elementos que estan en uno pero no en el otro
solo_clasicos = clasicos.difference(scifi)
solo_clasicos = clasicos - scifi
print(solo_clasicos)

# DIFERENCIA SIMETRICA -> ^ -> Elementos que estan en uno u otro conjunto pero no en ambos
unicos = clasicos.symmetric_difference(scifi)
unicos = clasicos ^ scifi
print(unicos)

# IsDisJoint -> Comprobar con un booleano si no tienen elementos en comun
print("¿Ciencia ficción y clásicos son unicas?", scifi.isdisjoint(clasicos))

# issubset -> issuperset -> Comprueba si un conjunto esta dentro de otro o lo contiene
miniset = {"Titanic", "El Padrino"}
print("¿El miniset esta dentro de clasicos?", miniset.issubset(clasicos))
print("¿Clasicos contiene el miniset?", clasicos.issuperset(miniset))
```
