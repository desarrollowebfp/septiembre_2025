## Tuplas


```python
# LAS TUPLAS SON INMUTABLES

heroe = ("Spiderman", "Peter Parker", 25, "Marvel")

print(len(heroe))
print(heroe.index("Peter Parker"))
print(heroe.count("Spiderman"))
print(heroe[0])
print(heroe[1:3])

# Para crear una tupla, mínimo tiene que haber una coma dentro de los parentesis, aunque una tupla de un solo elemento no tiene sentido, dado a que no podemos insertarle nada después porque es inmutable
nueva_tupla = ("Hola",)
print(type(nueva_tupla))

# Ejemplo de uso de tupla
config_servidor = ("localhost", 8080, True)
print(f"Conectado a {config_servidor[0]}:{config_servidor[1]}, SSL: {config_servidor[2]}")
```

```python
# Truco para poder modificar una tupla

config_servidor = ("localhost", 8080, True)
lista_config = list(config_servidor);
print(type(lista_config))
lista_config[0] = "Paco"
config_servidor = tuple(lista_config)


print(f"Conectado a {config_servidor[0]}:{config_servidor[1]}, SSL: {config_servidor[2]}")
```
